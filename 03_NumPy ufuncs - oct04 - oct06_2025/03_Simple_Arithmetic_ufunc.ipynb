#Simple Arithmetic
You could use arithmetic operators + - * / directly between NumPy arrays, but this section discusses an extension of the same where we have functions that can take any array-like objects e.g. lists, tuples etc. and perform arithmetic conditionally.


**Arithmetic Conditionally**: means that we can define conditions where the arithmetic operations should happen.

All of the dicussed arithmetic functions take a **where** parameter in which we can specify that condition.

```python
import numpy as np
```

#add()

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,9,8,7,6])

added_arr =np.add(arr1, arr2)
print(added_arr)
```

#subtract()

```python
subtracted_arr = np.subtract(arr1, arr2)
print(subtracted_arr)
```

#mutiply()

```python
multiplied_arr = np.multiply(arr1, arr2)
print(multiplied_arr)
```

#divide()

```python
divided_arr = np.divide(arr1, arr2)
print(divided_arr)
```

#powe()
rises the values from the first array to the power of the values of the second array.

```python
power_arr = np.power(arr1, arr2)
print(power_arr)
```

#Remainder - mod(), reminder()


```python
mod_arr = np.mod(arr1, arr2)
remainder_arr = np.remainder(arr1, arr2)

print(mod_arr)
print(remainder_arr)
```

#Quotient and Mod - **divmod()**
return both the quotient and mod. The return value is two arrays, the **first array contains the quotients** and the **second array contains the mod**.



```python
divmod_arr = np.divmod(arr1, arr2)
print(divmod_arr)
```

#Absolute values - absolute(), abs()

Both do the same absolute operation element wise but we should use the **absolute()** to avoid confusion with python's inbuilt **maths.abs()

```python
arr = np.array([-1, -2, 3, -4, -5])
absolute_arr = np.absolute(arr)
abs_arr = np.abs(arr)
print(absolute_arr)
print(abs_arr)
```
