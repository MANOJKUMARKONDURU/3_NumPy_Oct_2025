#Logs
NumPy provides functions to perform log at the base 2, e and 10.

We will also explore how we can take log for any base by creating a custom ufunc.

All of the log functions will place -inf or inf in the elements if the log can not be computed.



#Mathematically:

log2(N) = x means 2**x = N

expression | Meaning | Result

log2(1) => 2 ** 0 = 1 => 0

log2(2) => 2 ** 1 = 2 => 1

log2(4) => 2 ** 2 = 4 => 2

```python
import numpy as np
```

# Log at Base 2 - **log2()**
use the log2() function to perform log at the base 2.

---
Find log at base 2 of all elements of follwing array


```python
arr = np.arange(1, 10) #function returns an array with integers starting from 1(included) to 10(not included)
print(np.log2(arr))
```

#Log at Base 10 - **log10()**

```python
arr = np.arange(1, 11)
print(np.log10(arr))
```

#Natural Log, or Log at Base e -  **log()**
**loge(x) = y which means e ** y = x**

Expression | Meaning |  Result

log(1) => e ** 0 = 1 => 0

log(2) => e ** 0.6931 = 2 => 0.6931

log(3) => e ** 1.0986 = 3 => 1.0986

Where, e ~ 2.71828(Euler's number)

```python
arr = np.arange(1, 11)
print(np.log(arr))
```

# Log at Any Base

```python
from math import log
import numpy as np
```
