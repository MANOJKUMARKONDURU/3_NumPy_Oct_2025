#Rounding Decimals
There are primarily **Five** ways of rounding off the decimals in NumPy.
1. truncation - trunc()
2. fix - fix()
3. rounding - around()
4. floor - floor()
5. Ceil - ceil()

```python
import numpy as np
```

# Truncation, Fix - **trunc(), fix()**
Remove the decimals, and return the float number closest to the zero. Use the **trunc()** and **fix()** functions.


Will make the decimals as zero.

```python
arr = np.trunc( [-3.977, 3.1777] )
print(arr)


arr = np.fix( [-3.777, 3.1777] )
print(arr)
```

#Rounding - **around()**
The **around()** function rounds each element in an NumPy array to a specified number of decimal places.

#syntax
np.around(a, decimals = 0, out = None)
*  a = the input array
*  decimals = 0- round to the nearest integer. 2- round to two decimal places, -1 - round to nearest ten

* out = optional array to store the result



```python
#round to the nearest integer
arr = np.around([1.2, 2.47, 2.5, 2.51, 2.6])
print(f"round to the nearest integer: {arr}")

#round to two decimal places
arr = np.around([1.2, 2.47877, 2.5, 2.51, 2.677777], 2)
print(arr)

#Round to nearest Ten(Negative Decimal)
arr = np.around([11.2, 12.47877, 15.5, 22.51, 12.677777], -1)
print(arr)
```

#Floor - floor()
The **floor()** function rounds the decimal to **nearest lower integer**.


```python
arr = np.floor([1.2, 2.47, 2.5, 2.51, 2.9])
print(arr)
```

#Ceil - ceil()
The **ceil()** function rounds off the decimal to the **nearest upper integer**.

```python
arr = np.ceil([1.2, 3.57, 4.9])
print(arr)
```
