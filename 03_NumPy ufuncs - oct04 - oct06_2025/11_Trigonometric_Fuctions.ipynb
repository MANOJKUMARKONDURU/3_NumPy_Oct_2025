#Trigonometric Functions

NumPy provides the ufuncs **sin()**, **cos()** and **tan()** that take values the values in radians and produce the corresponding sin, cos and tan values.

```python
import numpy as np
```

```python
x = np.sin(np.pi/2)
print(x)
```

```python
arr = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])
x = np.sin(arr)
print(x)
```

#Convert Degrees into Radians
What Is a Radian?

A radian is the angle created when the length of an arc equals the radius of the circle. So:

*   If you wrap the radius of a circle around its edge, the angle that arc makes at the center is 1 radian.

*   A full circle is 2Ï€ radians, which equals 360 degrees.


By default all of the trigonometric functions take radians aas paremeters but we can convert radians to degrees and vice versa as well in NumPy


**Note**: Radians values are pi/180*degree_values.




#Degree to Radians

```python
arr = np.array([90, 180, 270, 360])
x = np.deg2rad(arr)
print(x)
```

#Radians to Degrees


```python
arr = np.array([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi, np.pi/4])
x = np.rad2deg(arr)
print(x)
```

#Finding Angles
Finding angles from values of sine,cos, tan.  E.g: sin, cos and tan inverse(arcsin, arccros, arctan)

NumPy provides ufuncs **arcsin(), arccos()** and **arctan()** that **produce radian values** for corresponding sin, cos, tan values give.


```python
x = np.arcsin(1.0)
print(x)

x = np.arccos(1.0)
print(x)

x = np.arctan(1.0)
print(x)
```

#Angles of each values in an Array


```python
arr = np.array([1, -1, 0.1])
x = np.arcsin(arr)

print(x)
```

#Hypotenues
Finding Hypotenus using pythagoras theorem in Numpy.

*    (Hypotenuse) ** 2 = (Base) ** 2 + (height) ** 2

NumPy provides the **hypot()** function that takes the base and perpendicular(height) values and produces the Hypotenuse based on Pythagoras theorem.

```python
base = 3
height = 4
x = np.hypot(base, height)
print(x)
```
