#How to create your own ufunc.
To create own ufunc, first you have to define a normal function in Pyhton, then you add it to the NumPy ufunc library with the **frompyfunc()** method.

The **frompyfunc()** method takes the following arguments:
1. **functions** - the name of the function.
2. **inputs** - the number of input arguments(arrays).
3. **outputs** - the number of ouput arrays.


```python
import numpy as np
```

```python
def myadd(x, y):
  return x + y

my_add = np.frompyfunc(myadd, 2, 1)

print(my_add([1, 2, 3, 4], [7, 5, 7, 8]))
```

#Check if a Function is a ufunc


```python
print(type(np.add))
print(type(my_add))
print(type(myadd))

print(type(np.concatenate)) #if it is not a ufunc then it will return another type, like this inbuit function

#this will give error because it doesn't exist.
#print(type(np.ajfkhaf))
```

To test if the function is a ufunc in an if statement, use the **numpy.ufunc** (or **np.ufunc** if you use as an alias for numpy).


```python
if type(np.add) == np.ufunc:
  print("add is a ufunc")
else:
  print("add is not a ufunc")
```
