#What are ufcns?

ufuncs stands for **Universal Functions"** and they are NumPy functions that operates on the **ndarray** object.


#Why use ufuncs?

ufuncs are used to implement **vectorization** in Numpy which is way faster than iterating over elements.
*   **Vectorization**: is the process of converting operations that would normally be done one element at a time(like loop) into operation that are applied to entire arrays or vectors at once.

ufuncs also takes additional arguments, like:

*    **where** - boolean array or condition definig where the operations should take place.
*   **dtype** - defining the returning type of elements.
*   **out** - ouput array where the return value should be copied.

#Add the elements of two lists

without ufuncs, we can use Python's built-in **zip()** method:


```python
x = [1, 2, 3]
y = [4, 5, 7]
z = []

for i, j in zip(x,y):
  z.append(i + j)
print(z)
```

NumPy has a ufunc for this, called **add(x, y)** that will produce the same result.

```python
import numpy as np
x = [1, 2, 3]
y = [3, 5, 7]
z = np.add(x, y)
print(z)
```
