# Differences - **diff()**
A discrete difference means substracting two successive elements.

E.g. for [1, 2, 3, 4], the discrete difference would be [2-1, 3-2, 4-3] = [1, 1, 1]

To find the discrete difference, use the **diff()** function.



```python
import numpy as np
```

#What **np.diff()** does
It computes:

*   newarr[i] = arr[i+1] -arr[i]

```python
arr = np.array([10, 15, 25, 5])
newarr = np.diff(arr)
print(newarr) #[5 10 -20] because 15-10=5, 25-15=10, and 5-25=-20
```

When n = 1

it returns => [5 10 -20] because 15-10=5, 25-15=10, and 5-25=-20

When n = 2

it returns => [5, -30] because 10-5= 5, -20-10= -30

```python
newarr = np.diff(arr, n = 2)
print(newarr)
```
