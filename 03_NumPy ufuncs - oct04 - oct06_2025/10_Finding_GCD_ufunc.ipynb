#Finding GCD(Greatest Common Divisior)
The GCD, also known as HCF(highest Common Factor) is the biggest number that is a common factor of both of the numbers.

```python
import numpy as np
```

```python
num1 = 8
num2 = 4
x = np.gcd(num1, num2)
print(x)
```

#Finding GCD in Arrays
To find the GCD of all the values in the array, you can use the **reduce()** method.

```python
arr = np.array([20, 8, 32, 36, 16])
x = np.gcd.reduce(arr)
print(x)
```
