#NumPy Data Types




*  **s** - strings

*  **i** - integer

*   **f** - float
*   **b** - boolean

*   **c** - complex float - eg: 1.0 + 2.0j
*  **m** - timedelta

* **M** - Datetime
* **O** - Object
* **S** - String
* **U** - unicode string
* **V** - fixed chunk of memory for other type(void)

---
Checking data type of an Array

```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4])
string = np.array(['aaaaaaa', 'b', 'c'])
print(arr.dtype)
print(string.dtype)
```

#Creating array with a Defined data type

```python
arr = np.array([1, 2, 3, 4], dtype='S')
print(arr)
print(arr.dtype)
```

#For i , u, f, S, U we can define size as well

```python
arr = np.array([1, 2, 3, 5], dtype = 'i4')
print(arr)
print(arr.dtype)
```

#Converting Data Type on Exsting Arrays
the best way to change the data type of an existing array, is to make a copy of the array with the **astype()** method

The **astype()** function creates a copy of the array and allows you to pass the data type as a parameter


```python
arr = np. array([0, 1, 2, 3.5 , 4.7])
integer = arr.astype("i")

decimals = arr.astype("f")
boolean = arr.astype(bool)

print(arr)
print(integer)
print(decimals)
print(boolean)
print(integer.dtype)
print(decimals.dtype)
print(boolean.dtype)
```
