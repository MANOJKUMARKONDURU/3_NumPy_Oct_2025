# Exponential Distribution
Is used for describing time till next event.
It has Two parameters.
1. **scale** - average waiting time(inverse of rate) - Default - 1.0

2. **size** - The shape of the returned array.



```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = random.exponential(scale = 2, size = (3, 4))
print(x)
```

```python
sns.displot(x, kind = 'kde')
plt.show()

sns.displot(random.exponential(size = 1000), kind = 'kde')
plt.show()
```

# Relation Between Poisson Distribution and Exponential Distribution.

Poisson Distribution deals with number of occurences of an even in a time period where as Exponential Distribution deals with the time between there events.
