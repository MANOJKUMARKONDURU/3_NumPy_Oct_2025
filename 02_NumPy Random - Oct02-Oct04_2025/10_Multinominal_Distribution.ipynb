# Multinomial Distribution

Multinominal distribution is a generalization of binomial distribution.

*   In binomial, each trail has only 2 outcomes(success/failure)
*   In Multinomial, each trail can have **more than 2 outcomes**.(eg. dice roll outcome)

It has three parameters.

1.  **n** - number of trails.
2.  **pvals** - list of probabilities of outcomes(eg: [1/6, 1/6, ...])
3.  **size** - The shape of the returned array



```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = random.multinomial(n = 12, pvals = [1/12, 1/12, 1/12, 1/12, 1/12, 1/12, 1/12, 1/12, 1/12, 1/12, 1/12, 1/12], size = (3))
print(x)
```

```python
sns.displot(random.multinomial(n = 3, pvals = [1/3, 1/3, 1/3], size = 3), kind = 'kde')
plt.show()
```

**Note:** Multinomial samples will not produce a single value. They will produce one value for each **pval**.
