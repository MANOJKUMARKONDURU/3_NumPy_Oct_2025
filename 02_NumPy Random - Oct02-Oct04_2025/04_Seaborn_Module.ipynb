#Seaborn

is a data visuvalization library that uses matplotlib to plot graphs. it will be used to visualize random distributions.








```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#Plotting a Displot(Distribution plot)

```python
sns.displot([0, 1, 2, 3])
plt.show()
```


#Plotting a Displot Without the Histogram



```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot([1, 2, 3], kind = "kde") # Kernal Density Estimate(KDE)
plt.show()
```
