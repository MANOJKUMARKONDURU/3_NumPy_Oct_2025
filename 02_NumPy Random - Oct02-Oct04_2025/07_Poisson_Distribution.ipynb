# Poisson Distribution
Poisson Distribution is a Discrete distribution.

*   it estimates how many times an event can happen in a specified time.

It has two parameters.
1.  **lam**(lambda) - rate or known number of occurences (expected number of events(mean))
2.  **size** - The shape of the returned array.

```python
import numpy as np

# Î» = 5 events on average
samples = np.random.poisson(lam=5, size=10)
print(samples)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
```

```python
x = random.poisson(lam = 2 , size = 1000)
print(x)
```

#Visuvalization of Poisson distribution


```python
sns.displot(random.poisson(x), kind= 'kde')
plt.show()

sns.displot(random.poisson(lam = 5, size = 1000), kind = "kde")
plt.show()
```

#Difference between Normal and Poisson Distribution
It will become similat to normal distribution with certain std dev and mean

```python
data = {
    "normal" : random.normal(loc = 50, scale = 7, size = 1000),
    "poisson" : random.poisson(lam = 50, size = 1000)
}
sns.displot(data, kind = "kde")
plt.show()
```

#Difference between Binomial and Poisson Distribution.

Binomial distribution has only two possible outcomes but poisson can have unlimited possible outcomes.

But for very large **n** and near-zero **p** binomial distribution is near identical to poisson distrubution such that n * p is nearly equal to **lam**

```python
data = {
    "Binomial" : random.binomial(n = 1000, p = 0.01, size = 1000),
    "Poisson": random.poisson(lam = 10, size = 1000)
}
sns.displot(data, kind = "kde")
plt.show()
```
