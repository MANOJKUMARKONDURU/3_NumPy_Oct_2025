# Normal Distribution
Gaussian Distribution - named aftered carl friedrich  gauss.

It fits the probability distribution of many events. eg: IQ scores, Heartbeats etc.


Use **random.normal()** method to get a Normal Data Distribution.

It has three parameters.

1.  **loc** - (Mean) Center of the normal distribution.


*   Default : 0.0


2.  **scale** - (standard Deviation - spread or width of the normal distribution) how flat the graph distribution should be.
*   Default: 1.0  
3.  **size** - The shape of the returned array.



# Generate a random distribution of size =2*3

```python
from numpy import random
```

```python
x = random.normal(size = (2, 3))
print(x)
```

# Visualization of Normal Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.random.normal(size = 100)
sns.displot(x, kind = "kde")
plt.show()
```

# The curve of Normal Distribution is also known as the Bell Curve because the bell-shaped curve

```python

arr = np.random.normal(loc = 10, scale = 2, size = 100)
print(arr)
```

```python
sns.displot(arr, kind = 'kde')
plt.show()
```
