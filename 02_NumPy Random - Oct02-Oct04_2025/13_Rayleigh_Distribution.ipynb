#Rayleigh Distribution
Rayleigh used in signal processing.


It has two parameters.
1.  **scale** - (standard deviation) decides how flat the distribution will be. default - 1.0

2. **size** - the shape of the returned array.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = random.rayleigh(scale = 3, size = (2, 3))
print(x)
```

```python
sns.displot(random.rayleigh(size = 1000), kind = 'kde')
plt.show()
```

#Similarity between Rayleigh and Chi Square Distribution.
At Unit stddev and 2 degree of freedom rayleigh and chisquare represents the same distribution.

```python
data ={
    'rayleigh': random.rayleigh(size = 1000),
    'chisquare': random.chisquare(df = 2, size = 1000)
}

sns.displot(data, kind = 'kde')
plt.show()
```
