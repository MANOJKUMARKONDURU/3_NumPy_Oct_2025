# Random Permutation

Permutation means an ordered arrangement of elements where the sequence matters.
Eg: [3, 2, 1] is a permutation of [1, 2, 3] and vice-versa.

The NumPy Random Module provides two methods for this : **shuffle()** and **permutation()**


```python
from numpy import random
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)
print(arr)
```

#**shuffle()** methods makes changes to the original array

# Generating Permutation of Arrays


```python
arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
```

#The **permutation() method return a re-arranged array and leaves the original array unchanged
