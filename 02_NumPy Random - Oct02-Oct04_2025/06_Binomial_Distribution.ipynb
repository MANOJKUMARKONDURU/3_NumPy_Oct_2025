#Binomial Distribution

Binomial Distribution is a Discrete Distribution.

*   A discrete distribution describes outcomes that take distinct, seperate values.
*   Binomial Distribution = probability of getting "k successes" in 'n trials' when each trails has 2 outcomes and the same chances of success.

It has three parameters:
1.  **n** - number of trails
2.   **p** - probability of occurence of each trail(eg. for toss of a coin 0.5 each)
3.   **size** - The sape of the array


```python
from numpy import random
```

```python
x = random.binomial(n = 10, p = 0.5, size = 1000)
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(x)
plt.show()
```

# difference between Normal and Binomial distribution

The main difference is that normal distribution is continuous where binomial is discrete, if there is enough data points it will be quite similar to normal distribution with certain loc and  scale.


```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "normal" : random.normal(loc = 50, scale = 5, size = 1000),
    "binomial" : random.binomial(n = 100, p = 0.5, size = 1000)
}

sns.displot(data = data, kind = "kde")
plt.show()
```
