#Uniform Distribution

Used to descibe probability where **every event has equal chances of occuring**.

Eg: Generation of random numbers.

It has three parameters.
1.  **low** - lower bound - default 0.0
2. **high** - upper bound - default 1.0
3. **size** - The shape of the returned array.

```python
from numpy import random
```

Create 2*3 uniform distribution sample


```python
x = random.uniform(size = (2, 3))
print(x)
```

#Visuvalization of Uniform Distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
x = random.uniform(size = 1000)
sns.displot(x, kind = 'kde')
plt.show()
```
