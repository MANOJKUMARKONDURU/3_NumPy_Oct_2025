#Chi Square Distribution

Chi Square Distribution is used as a basis to verify the hypothesis(to verify the assumption).

It has two parameters.

1.  **df** - degree of freedom.(How many independent pieces of informations(or variables) are used to build a distribution. It's a parameter that shapes the distribution)

2. **size** - The shape of the returned array.


```python
from numpy import random
```

```python
x = random.chisquare(df = 2, size = (2, 3))
print(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sns.displot(random.chisquare(df = 2, size = 1000), kind = 'kde')
plt.show()
```
