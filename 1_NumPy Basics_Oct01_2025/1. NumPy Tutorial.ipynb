1 - D Like a list  
2 - D like a matrix  
3 - D like a cube ... so on

Array means collecction of items(usually numbers) stored in grid-like structure   
**All elements must be same data type**   
It has attributes like


*   .ndim - number of dimensions
*   .nshape - size along each dimension
*   .dtype - data type of an element

# Why NumPy faster than Lists?
NumPy arrays are stored in one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.


This behaviour is called locality of reference in computer science.



```python
import numpy
arr = numpy.array([1, 2, 3, 4])
print(arr)
```

#1-D Arrays
An array that has 0-D arrays as it's elements is called 1-D array


```python
arr = np.array([1, 2, 3, 4, 7])
print(arr)
```

#2-D Arrays
An array that has 1-D arrays as it's elements is called 2-D array

Often used to represent matrix

**Arrays that have rows and columns**

**Shape -  [row],[column]

```python
arr = np.array([[1, 2, 3], [4, 5, 7]])

print(arr)
print(arr.ndim)
```

```python
arr = np.array([ [1, 2], [2, 3], [4, 5]])

print(arr)
print(arr.ndim)
```

#3-D Arrays
An array that has 2-D(matrices) as it's elements called 3-D array

**Arrays that have depth, rows, columns**

**Shape - [deptth][row][column]






```python
arr = np.array([
     [
        [1], [2]
    ],
    [
        [3], [4]
    ]
])
print(arr)
print(arr.ndim)
```

```python
arr = np.array([[[1, 2, 3], [4, 5, 7]],[[1, 4, 7], [2, 3, 7]]])
print(arr)

print(arr.ndim)
```

#Higher dimensions

An array can have any number of dimensions

You can define the number of dimensions by using **ndmin**








```python
arr = np.array([1,2, 3, 4, 5], ndmin = 5)
print(arr)
print(arr.ndim)
```
