#Sorting Arrays  - **sort()**

Sorting means putting elements in an ordered sequence

Ordered sequence means its any sequence of order corresponding to elements, like numeric or aplabetical, ascending or descending.

The NumPy ndarray object has a funcion called **sort()**, that will sort a specified array.


```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 5, 4, 7, 3])
arr = np.sort(arr1)
print(arr)
```

**This method return a copy of the array, leaving the original array unnchanged**

You can also sort arrays of strings or any other dara types

```python
arr_string = np.array(["Mango", "Lambo", "Mustang"])

print(np.sort(arr_string))
```

#sort a boolean array

```python
boolean_arr = np.array([True, False, False, True])

print(np.sort(boolean_arr))
```

#Sorting a 2-D Array
Both arrays will be sorted

```python
arr = np.array([[3, 2, 1], [7, 5, 4]])
print(np.sort(arr))
```

# Sortin 3-D array




```python
arr3 = np.array([[[1, 21, 2], [7, 3, 5]], [[27, 17, 15], [100, 97, 77]]])

print(np.sort(arr3))
```
