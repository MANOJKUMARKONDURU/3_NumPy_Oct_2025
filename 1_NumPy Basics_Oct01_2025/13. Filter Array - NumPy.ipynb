#Filtering Arrays
Getting some elements out of an existing array and creating a new array out of them is called filtering.

in NumPy, you can filer an array using a boolean index list.


If the value at an index is **True** that element is contained in filtered array, if the value is at the index **False** that element is excluded from the filtered array.


```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,4,5])

x = [True, False, True, False, True]

newarr = arr1[x]

print(newarr)
```

#Creating the Filter Array

create a filter array based on conditions.

```python
arr = np.array([1, 2, 4, 7, 21, 51, 34, 54, 90])

# Create an empty list
filter_arr = []

# Traverse through the array
for element in arr:
  if element > 20:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr = arr[filter_arr]


print("Filter array: ", filter_arr)
print("New array: ", newarr)
```

# Return only  Even



```python
arr = np.array([1, 2, 4, 7, 21, 51, 34, 54, 90])

filter_arr = arr % 2 == 0

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```
