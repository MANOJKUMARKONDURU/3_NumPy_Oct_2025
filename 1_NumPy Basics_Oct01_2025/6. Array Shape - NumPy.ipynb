#Shape of an Array

is the number of elements in each dimension.

NumPy has an attribute called **shape** that returns a tuple with each index having the number of correspoding elements



```python
import numpy as np
```

```python
# 1-D Array
arr = np.array([1, 2, 3, 4])
print(arr.shape)


# 2-D Array
arr2 = np.array([
    [1, 2, 3],
     [4, 5, 7]
    ])
print(arr2.shape)  # (rows, columns)


# 3- D Arrays
arr3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 7], [8, 9]]
])

print(arr3.shape) # (dimensions, row, column)

#5-D array
arr5 = np.array([1, 2, 3], ndmin = 5)
print(arr5.shape)
```
