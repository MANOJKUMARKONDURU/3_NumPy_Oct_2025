#Reshaping Arrays

Means changing the shape of the array with **reshape()**.

By reshaping we can add or remove dimensions or change the number of elements in each dimension.
---
#Reshape From 1-D to 2-D, 1-D to 3-D

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5, 6, 7, 8])

arr2 = arr1.reshape(2, 4)
arr3 = arr1.reshape(2, 2, 2)
print(arr2)
print()
print(arr3)
```

#Reshape results view

```python
print(arr2.base)

print(arr3.base)
```

#Unknown Dimension

Allowed to have one "Unknown dimension"


Pass **-1** as the value and NumPy will calculate for you

```python
new_arr = arr1.reshape(2, 2, -1)
print(new_arr)
print()

new_arr2 = arr2.reshape(-1, 2, 2)
print(new_arr2)
print()
new_arr3 = new_arr2.reshape(2, -1)
print(new_arr3)
```

#Flattening the Arras
Means converting a multidimensional array into 1-D array

we can use **reshape(-1) to do this

```python
new_arr1 = new_arr3.reshape(-1)
print(new_arr1)
```
