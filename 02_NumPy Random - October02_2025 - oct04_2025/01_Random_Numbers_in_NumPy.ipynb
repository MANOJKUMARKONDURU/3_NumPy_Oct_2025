#What is a Random Number?

Random number does **not** mean different number every time. Random number means which cannot be predicted Logically.

#Pseudo Random and True Random

Computers work on programs, and programs are definitive set of instructions. So there should be some algorithm to generate a random number as well.

**If there is a program to generate random number that can be predicted, thus its not a random number**

Random numbers generated through generation algorithm are called **pseudo random**

**Can we make truly random numbers? - Yes**

Yes. In order to generate truly random on our computers we need to get the random data from some outside source. This outside source is generally our keystrokes, mouse movements, data on network etc.

**We don't need truly random numbers, unless it is realated to security(eg: encryption keys)**

#Generate Random Number

NumPy offers the **random** module to work with randdom numbers.

```python
from numpy import random
```

```python
# Generate Random integer from 0 to 100

x = random.randint(100)
print(x)
```

# Generate Random Float

the random module's **rand()** method returns a random float between 0 and 1.


```python
x = random.rand()
print(x)
```

# Generate Randoom Array
You can use the two methods from the above examples to make random arrays.
#Integers
The **randint()** method takes a **size** parameter where you can specify the shape of an array.


```python
x = random.randint(15, size = (3))  # size = (how many elemnts)
print(x)
print()

x = random.randint(100, size = (3, 5))  #size = (how many rows, how many columns)
print(x)

print("3-D array")
x = random.randint(100, size = (2, 3, 5)) #size = (dimensions, rows, columns)
print(x)
```

# Float

The **rand()** method allows to specify the shape of the array






```python
x = random.rand(5)
print(x)

print("\n2-D array with rand()")
x = random.rand(3, 3)
print(x)
print("\n3-d array with rand()")
x = random.rand(2, 3, 5)
print(x)
```

# Generate Random number from an Array

The **choice()** method allows to generate a random number based on the array of values.

It takes array as a parameter and randomly returns one of the values.

To print specified number of values or shape of the array we can use **size**

```python
x = random.choice([1, 2, 3, 4, 5])
print(x)

print("\n2-D Array")
x = random.choice([1, 2, 3, 4, 5], size = (3, 5))
print(x)

print("\n3-d Array")
x = random.choice([1, 2, 3, 4, 5], size = (2, 3, 5))
print(x)
```
